```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
Bcoin = pd.read_csv('./BitCoin.csv', index_col="Date", parse_dates=True, usecols=["Date", "Close"])
Bcoin.sort_index(inplace=True)
Bcoin.head()
```

```python
Bcoin = Bcoin['2016-06-01':'2017-06-30']
```

```python
Bcoin
```

```python
ma_5d = Bcoin.Close.rolling(window=5).mean()
```

```python
ma_5d
```

```python
plt.figure(figsize=(15,10))

plt.xlabel('date')
plt.ylabel('price')

plt.xticks(rotation=45)

plt.plot(ma_5d, '#f2a900')
plt.title("BitCoin")
plt.show()
```
