## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리
- 프레임 워크 
- Matplotlib 의 라이브러리
- % matplotlib inline 을 통해 활성화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study

```python
plt.plot([2,4,2,4,2])
plt.show()
```

```python
plt.figure(figsize=(6,6)) #plotting 을 할 도면을 설정\
plt.plot([2,4,2,4,2])
plt.plot([1,2,3,4,5])
plt.show()
```

```python
### 2차 함수 그리기
x = np.arange(-10,10,0.01)
y = x**2
# x,y 축설명
plt.xlabel('x value')
plt.ylabel('f(x) value')

# x 축의 범위를 지정
plt.axis([-5,5,0,25])
# x축과 y축의 범위를 설정해주자
plt.xticks([i for i in range(-5,6 ,1)])
plt.yticks([i for i in range(0,27,3)])
plt.plot(x,y, label = 'trend')

# title
plt.title('y = x^w graph')

# 범례 legend
plt.legend()
plt.show()
```

## III. Matplotlib Case Study

```python
x = np.arange(20)
y = np.random.randint(0,20,20) # 난수를 20번 생성
plt.plot(x,y)
# Extra : y축을 20까지 보이게 하고 싶다면.!?
plt.axis([0,20,0,20])
plt.yticks([0,5,10,15,20])
plt.show()
```

### Scatter plot

```python
 plt.scatter(x,y)
```

### boxplot 전반적인 분포를 나타내줌
- 수치형 데이터에 대한 정보 (q1,q2,q3,q4)

```python
plt.boxplot((x,y))
# Extra : plot 의 title 을 지정
plt.title('Bow plot x & y')
plt.show()
```

### 막대 그래프 
- 범주형 데이터 의 '값' 과 '크기'를 직사각형으로 나타낸 그림

```python
plt.bar(x,y)
plt.xticks(np.arange(0,20,1))
# Exta: xticks 를 올바르게 처리해보자
plt.show()
```

### Hsitogram
- 도수 분표를 막대 형태로 나타낸다.
- '계급' 이 존재

```python
plt.hist(y, bins = np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

### Piechart
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그림
- 다름 그래프에 비해서 비율을 판단하기에 편함

```python
plt.pie(y , labels = np.arange(0,20))
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### MATPLOTLIB을 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널 밀도 그림
- 카운트 그림
- 캣 그림
- 스트립 그림
- 히트맵

```python
import seaborn as sns
```

```python
x = np.arange(0,22,2)
y = np.random.randint(0,20,20)
```

### 커널 밀도 그림(kernel Density Plot)
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그림 그림
- sns.kdeplot()

```python
sns.kdeplot(y, shade = True)
```

### Countplot 카운트 그림
- 범주형 column의 변도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- sns.countplot()

```python
vote_df = pd.DataFrame({'name':['Andy','Bob','Cat'],'Vote':[True,False,True]})
vote_df
```

```python
vote_count = vote_df.groupby('Vote').count()
vote_count
```

```python
plt.bar(x=[False,True],height = vote_count['name'])
```

```python
sns.countplot(x = vote_df['Vote'])
```

### Cat 그림 (Cat plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- sns.catplot()

```python
covid = pd.read_csv('../kaggle/archive/country_wise_latest.csv')
covid.head()
```

```python
s= sns.catplot(x = 'WHO Region', y = 'Confirmed', data=covid)
s.fig.set_size_inches(10,6)
```

### 스트립 그림(Strip plot)
- scatter plot 과 유사하게 데이터의 수치를 표현하는 그래프
- sns.stripplot()

```python
sns.stripplot(x = 'WHO Region', y='Recovered', data =covid)
```

### Heat map
- corr() 을 제일 많이 사용
- sns.heatmap()

```python
sns.heatmap(covid.corr())
```
