## Extra Mission

1.캐릭터는 저마다 지능, 힘, ... 등 다양한 수치를 지니고 있다.   
또한 각 캐릭터는 DC, 마블 등 다양한 코믹스 회사를 바탕으로 하고 있다.   
어떤 코믹스 회사의 캐릭터들이 능력치 합의 평균이 가장 큰가? 이를 보이기 위한 과정을 보여라.

```python
# 코믹스 회사와 캐릭터를 매치해주어야 한다. 
# 캐릭터의 코믹스 회사 정보가 담긴 파일을 불러오자.

comics = pd.read_csv('marvel_characters_info.csv')
comics.head()
```

```python
comics.Publisher.unique()
```

```python
# 두 데이터프레임을 공통 열 'Name'을 기준으로 합치자

df_merged = pd.merge(abengers, comics)
df_merged.head()
```

```python
# 코믹스 회사별로 캐릭터들의 능력치 총합의 평균을 구해보자

spec_mean_by_publisher = pd.DataFrame(df_merged.groupby(['Publisher']).Total.mean()).reset_index()
```

```python
# 능력치 총합의 평균을 기준으로 내림차순 정렬

spec_mean_by_publisher.sort_values(by='Total', axis=0, ascending=False)

## 'Shueisha' 코믹스 회사의 캐릭터 능력치 총합이 큰 것으로 보인다. 
```

2.좋은 캐릭터와 나쁜 캐릭터가 격돌한다고 한다. 격돌하는 경우 캐릭터들의 능력치의 합의 평균이 큰 팀이 이긴다고 한다.  
단, 불의를 못참는 중립 캐릭터들은 중립캐릭터가 없었을 당시에 열세인 팀에 가담한다.  
이러한 상황일때 결과적으로 어떤 캐릭터 진영이 승리할 것인가? 이를 보이기 위한 과정을 보여라.

```python
# 좋은 캐릭터들의 능력치 합의 평균
abengers[abengers['Alignment'] == 'good'].Total.mean()
```

```python
# 나쁜 캐릭터들의 능력치 합의 평균
abengers[abengers['Alignment'] == 'bad'].Total.mean()

## 중립 캐릭터들은 열세인 좋은 캐릭터 진영에 가담한다.
```

```python
# 좋은 캐릭터들 + 중립 캐릭터들의 능력치 합의 평균
abengers[(abengers['Alignment'] == 'good') | (abengers['Alignment'] == 'neutral')].Total.mean()

## 그래도 나쁜 캐릭터들의 능력치 합의 평균보다 낮다...
## 나쁜 캐릭터 진영이 승리할 것...
```
