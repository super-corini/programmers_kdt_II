# 전세계 covid-19 백신 추세
[Covid-19 World Vaccination Progress](https://www.kaggle.com/gpreda/covid-world-vaccination-progress?select=country_vaccinations.csv)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('./country_vaccinations.csv')
df.head(1)
```

```python
df.describe()
```

## 한국 백신 접종 상황

```python
kor_df = df[df['iso_code']=='KOR']
```

```python
kor_df.tail(3)
```

```python
plt.figure(figsize=(10,5))
plt.style.use("ggplot")
sns.lineplot(x="date", y="people_vaccinated", data=kor_df)

plt.xticks([i for i in range(0, 75, 5)],rotation=45)
plt.xlabel("Date")
plt.ylabel("Vaccinated")

plt.show()
```

vaccinated per hundred

```python
plt.figure(figsize=(10,5))
plt.style.use("ggplot")
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=kor_df)

plt.xticks([i for i in range(0, 75, 5)],rotation=45)
plt.yticks([i for i in range(0, 110, 10)])
plt.xlabel("Date")
plt.ylabel("Vaccinated per hundred")

plt.show()
```

### 다른 나라와 비교

```python
us_df = df[df['iso_code']=='USA']
us_df.head(2)
```

```python
plt.figure(figsize=(10,5))
plt.style.use("ggplot")
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=us_df)
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=kor_df)

plt.xticks([i for i in range(0, 150, 5)],rotation=45)
plt.yticks([i for i in range(0, 110, 10)])
plt.xlabel("Date")
plt.ylabel("Vaccinated per hundred")

plt.legend(['US', 'KOR'])

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.style.use("ggplot")
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=us_df)
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=kor_df)
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=df[df['iso_code']=='JPN'])
sns.lineplot(x="date", y="people_vaccinated_per_hundred", data=df[df['iso_code']=='DEU'])

plt.xticks([i for i in range(0, 150, 5)],rotation=45)
plt.yticks([i for i in range(0, 110, 10)])
plt.xlabel("Date")
plt.ylabel("Vaccinated per hundred")
plt.title("World Vaccinated Progress ")

plt.legend(['US', 'KOR', 'JPN', 'DEU'])

plt.show()
```

```python
x = df.groupby("country")['total_vaccinations_per_hundred'].mean().sort_values(ascending=False).head(50)
plt.figure(figsize=(5, 20))
ax = sns.barplot(x.values, x.index)
ax.set_ylabel("Country")
ax.set_xlabel("Vaccinations per hundred")
plt.title("World Vaccinations per hundred Rank")

plt.show()
```
