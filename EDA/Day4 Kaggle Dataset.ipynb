### 지표로 사용할 Column만 분리

```python
lck = lol_esports[lol_esports['league'] == 'LCK'][['patch', 't1_playerid', 't2_playerid', 'game', 'gamelength', 't1_result',
     't1_kills', 't1_deaths', 't1_assists', 't1_firstblood', 't1_firsttower', 't1_dragons', 't1_heralds', 't1_barons', 
     't1_towers', 't1_dpm','t1_damagetakenperminute', 't1_visionscore', 't1_totalgold','t1_goldat15', 't1_xpdiffat15', 't1_golddiffat15', 
      't2_result','t2_kills', 't2_deaths', 't2_assists', 't2_firstblood', 't2_firsttower', 't2_dragons', 't2_heralds', 't2_barons', 
     't2_towers', 't2_dpm','t2_damagetakenperminute', 't2_visionscore', 't2_totalgold','t2_goldat25', 't2_xpdiffat25', 't2_golddiffat25',]]
lck['patch'] = lck['patch'].astype(str)
lck
```

## 분석의 목적과 변수 확인

- patch : 해당 경기에서 사용된 클라이언트 패치 버전
- t1_playerid, t2_playerid, game, gamelength : 블루팀, 레드팀, 매치 내 게임 번호, 경기 시간 
- result : 각 팀의 승패 결과
- kills, deaths, assists, towers : 각 팀의 킬, 데스, 어시스트, 타워 파괴
- firstblood, firsttower : 선취 킬, 타워 파괴 여부
- dragons, heralds, barons : 각 팀의 오브젝트 획득 수
- dpm, damagetakenperminute : 각 팀의 분당 가한 데미지, 분당 받은 데미지
- visionscore : 각 팀의 시야 점수
- xpdiffat15, xpdiffat25 : 15분 경험치 차이, 25분 경험치 차이
- totalgold, goldat15, golddiffat15, t2_goldat25, t2_golddiffat25 : 각 팀의 전체 골드, 15분, 25분 골드 차이

## 유의미한 Feature 찾기

### kills + deaths + assists -> result
- KDA와 승패의 상관성

```python
lck['t1_kda'] = (lck['t1_kills'] + lck['t1_assists']) / lck['t1_deaths']
lck['t2_kda'] = (lck['t2_kills'] + lck['t2_assists']) / lck['t2_deaths']
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
sns.stripplot(x='t1_result', y='t1_kda', data=lck, ax=ax)
sns.stripplot(x='t2_result', y='t2_kda', data=lck, ax=ax)

plt.xlabel('result')
plt.ylabel('kda')

plt.title('KDA -> Result')
plt.show()
```

### dragons + heralds + barons -> result
  - 오브젝트 획득과 승패의 상관성

```python
lck[['t1_result', 't1_barons', 't1_heralds', 't1_dragons']].groupby(['t1_result']).mean()
```

```python
lck['t1_totalobject'] = lck['t1_dragons'] + lck['t1_heralds'] + lck['t1_barons']
lck['t2_totalobject'] = lck['t2_dragons'] + lck['t2_heralds'] + lck['t2_barons']
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
sns.stripplot(x='t1_result', y='t1_totalobject', data=lck, ax=ax)
sns.stripplot(x='t2_result', y='t2_totalobject', data=lck, ax=ax)

plt.xlabel('result')
plt.ylabel('totalobject')

plt.title('totalobject -> Result')

plt.show()
```

### firstblood + firsttower -> result
  - 선취 득점과 승패의 상관성
  - 퍼스트 블러드를 따냈을 때의 승률 : 69.9%
  - 퍼스트 타워를 따냈을 때의 승률 : 79.8%

```python
first_win = pd.concat([lck[['t1_result', 't1_firstblood']].groupby(['t1_firstblood']).mean(), lck[['t1_result', 't1_firsttower']].groupby(['t1_firsttower']).mean()], 
                      keys=['t1_firstblood', 't1_firsttower'], names=['category', 'score'])
first_win
```

```python
plt.figure(figsize=(10,6))

ax1= plt.subplot(121)
plt.ylim(0, 1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.title('First Blood -> Result')
lck[['t1_result', 't1_firstblood']].groupby(['t1_firstblood']).mean().plot(kind='bar', ax=ax1, xticks=[i for i in range(2)], rot=0)

ax2 = plt.subplot(122, sharey= ax1)
plt.title('First Tower -> Result')
lck[['t1_result', 't1_firsttower']].groupby(['t1_firsttower']).mean().plot(kind='bar', ax=ax2, legend=None, xticks=[i for i in range(2)], rot=0)

plt.show()
```

### gold -> result
- 골드 획득과 승패의 상관성

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

sns.kdeplot(x = lck[lck['t1_result'] == 1]['t1_totalgold'], ax = ax)
sns.kdeplot(x = lck[lck['t1_result'] == 0]['t1_totalgold'], ax = ax)

plt.legend(['Victory', 'Defeat'])

plt.title('Gold -> Result')
plt.show()
```
