## 각 진영 캐릭터의 능력치 평균을 알아보고 plot

```python
ab_stats_df.groupby(['Alignment']).mean()
```

```python
plt.plot(ab_stats_df.groupby(['Alignment']).mean().transpose())
plt.legend(['bad','good','neutral'])
```

평균적인 능력치는 중립진영에서 가장높았지만 캐릭터 수가 압도적으로 작습니다.  
good, bad에서도 평균적인 능력치는 bad가 높습니다.

```python
sns.catplot(x='Total',col='Alignment',data=ab_stats_df)
sns.catplot(x='Power',col='Alignment',data=ab_stats_df) # 각 지표를 비교해서 확인 할 수 있다.
```
