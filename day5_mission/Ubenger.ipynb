#### 나쁜 캐릭터와 좋은 캐릭터의 능력치 분포

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
d=pd.read_csv("./archive/charcters_stats.csv")

che=(d['Alignment']!='neutral')

d=d[che]
for i in d.iloc[:,2:-1]:
    sns.boxplot(x="Alignment",y=i,data=d,hue="Alignment")
    plt.show()

    
```
