## 좋은 캐릭터 vs 나쁜 캐릭터의 능력치들의 수치분포

```python
heros_count = heros_df.groupby(by="Alignment").count()
heros_count
```

```python
plt.bar(np.array(heros_count.index), heros_count["Name"])
plt.show()
```

```python
heros_mean = heros_df.groupby(by="Alignment").mean()
heros_mean
```

```python
plt.bar(np.array(heros_mean.index), heros_mean['Total'])
plt.show()
```
