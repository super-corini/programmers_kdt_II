### 6. 눈 색의 분포

```python
chracter_info[chracter_info.EyeColor == '-'] = 'No Data'
eye_color_data = chracter_info.groupby('EyeColor')['ID'].count().sort_values(ascending=False)
eye_color_data
```

```python
etc = pd.Series({"Etc":sum(eye_color_data[6:])})
eye_color_data = eye_color_data[:6].append(etc).sort_values(ascending=False)
eye_color_data
```

```python
explode = [0.12 for _ in range(len(eye_color_data))]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0', '#fff8f8', '#fff8d8', '#ffefef']
plt.pie(eye_color_data, labels=eye_color_data.index, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.title('Eye Color', fontsize=15)
plt.show()
```

### 7. 강함 = 등장횟수? 

### 중복작품을 포함한, 모든 작품에서의 등장횟수

```python
pop_data = filmography_data.groupby('name')['comicID'].count().sort_values(ascending=False)
merge_data = pd.merge(stats_data, pop_data, left_on='Name', right_on='name')
merge_data.rename(columns={'comicID':'Number of appearances'}, inplace=True)
merge_data
```

```python
corr = merge_data.corr()
corr
```

```python
plt.scatter(merge_data['Number of appearances'], merge_data['Total'])
```

### result : 강한 캐릭터라고 등장횟수가 많은 것은 아니다.
