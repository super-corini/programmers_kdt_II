```python
sns.countplot(x='Alignment',data=stats_df)
plt.show()
```

```python
stats_df=stats_df.dropna(subset=['Alignment'])
stats_df.info()
```

```python
f, ax = plt.subplots(4,2, figsize=(16,48))
sns.boxplot(x='Alignment',y='Intelligence', data=stats_df, ax=ax[0][0])
sns.scatterplot(x='Alignment',y='Intelligence', data=stats_df, ax=ax[0][0])
ax[0][0].set_title('Intelligence')
sns.boxplot(x='Alignment',y='Strength', data=stats_df, ax=ax[0][1])
sns.scatterplot(x='Alignment',y='Strength', data=stats_df, ax=ax[0][1])
ax[0][1].set_title('Strength')
sns.boxplot(x='Alignment',y='Speed', data=stats_df, ax=ax[1][0])
sns.scatterplot(x='Alignment',y='Speed', data=stats_df, ax=ax[1][0])
ax[1][0].set_title('Speed')
sns.boxplot(x='Alignment',y='Durability', data=stats_df, ax=ax[1][1])
sns.scatterplot(x='Alignment',y='Durability', data=stats_df, ax=ax[1][1])
ax[1][1].set_title('Durability')
sns.boxplot(x='Alignment',y='Power', data=stats_df, ax=ax[2][0])
sns.scatterplot(x='Alignment',y='Power', data=stats_df, ax=ax[2][0])
ax[2][0].set_title('Power')
sns.boxplot(x='Alignment',y='Combat', data=stats_df, ax=ax[2][1])
sns.scatterplot(x='Alignment',y='Combat', data=stats_df, ax=ax[2][1])
ax[2][1].set_title('Combat')
sns.boxplot(x='Alignment',y='Total', data=stats_df, ax=ax[3][0])
sns.scatterplot(x='Alignment',y='Total', data=stats_df, ax=ax[3][0])
ax[3][0].set_title('Total')
plt.show()
```
