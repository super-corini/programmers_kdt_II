```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## II. Matplotlib Case Study with Arguments

```python
# 꺾은선 그래프

plt.plot([1, 2, 3, 4, 5]) # 실제 plotting 을 하는 함수, y = x + 1, x = index, y = [1, 2, 3, 4, 5]

plt.show() # plt 를 확인하는 명령
```

## Figsize : Figure(도면) 의 크기 조절

```python
#plt 의 사이즈를 조정

plt.figure(figsize=(3, 3)) # plotting 을 할 도면을 선언, 1 pix = 72 -> (6, 6) = (72*6, 72*6)

plt.plot([0, 1, 2, 3, 4])

plt.show()
```

### 2차함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y=x 를 그려보면:

plt.plot([0, 1, 2, 3, 4])

plt.show()
```

```python
# numpy.array 를 이용해서 함수 그래프 그리기
# y = x^2

x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x)

plt.plot(x,y)

plt.show()

# 스무스 하지 못하다는 단점이 있음.
```

```python
# np.arange()
# range 와 다른 것은 시작점, 끝점, 증감값을 적어줄 수 있음

x = np.arange(-10, 10, 0.01)
x[:5]
```

```python
plt.plot(x, x**2)

plt.show()
```

```python
# x, y 축에 설명 추가하기

x = np.arange(-10, 10, 0.01)

###
plt.xlabel("x value") # x 축의 이름을 결정
plt.ylabel("f(x) value") # y 축의 이름을 결정
###

plt.plot(x, x**2)

plt.show()
```

```python
# x, y 축의 범위를 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value") 
plt.ylabel("f(x) value")

###
plt.axis([-5, 5, 0, 25]) # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)

plt.show()
```

```python
# x, y 축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value") 
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) 

###
plt.xticks([i for i in range(-5,6,1)]) # x 축의 눈금 설정: -5, -4, -3, ...
plt.yticks([i for i in range(0,27,3)]) # y 축의 눈금 설정
###

plt.plot(x, x**2)

plt.show()
```

```python
# 그래프에 title 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value") 
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) 
plt.xticks([i for i in range(-5,6,1)]) 
plt.yticks([i for i in range(0,27,3)])

###
plt.title("y = x^2 graph")
###

plt.plot(x, x**2)

plt.show()
```

```python
# 그래프에 범례 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value") 
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) 
plt.xticks([i for i in range(-5,6,1)]) 
plt.yticks([i for i in range(0,27,3)])
plt.title("y = x^2 graph")

###
plt.plot(x, x**2, label = "trend")
plt.legend()
###

plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프 (Plot)
<br/>

- <code>.plot()</code>
- 시계열 데이터에서 많이 사용
- 이전에 비해서 이후의 값이 얼마나 변했는지 확인해주는 용도로 사용 가능
- 다른 그래프에 비해서 데이터의 **분포** 확인에 용이

```python
x = np.arange(20) # 0 ~ 19
y = np.random.randint(0, 20, 20) # 0~20, 난수를 20 번 생성

x, y
```

```python
plt.plot(x, y)

plt.show()
```

```python
# Extra : y 축을 20 까지 보이게 하고싶다면?, y 축을 "5" 단위로 보이게하고 싶다면?
# .axis(), .ytics()
plt.plot(x, y)

plt.axis([0, 20, 0, 20])
plt.yticks([0, 5, 10, 15, 20])

plt.show()
```

### 산점도 (Scatter Plot)
<br/>

- <code>.scatter()</code>
- 두 변수가 독립적일 때 자주 사용
- 변수 사이의 상관관계를 파악 가능
- 다른 그래프에 비해서 데이터의 **분포** 확인에 용이

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)
<br/>

- <code>.boxplot()</code>
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)
- 단일변수가 있을 때, 변수의 전반적인 분포가 어떤지 확인하고 싶을 때 자주 사용
- 다른 그래프에 비해서 데이터의 **축약된 정보** 확인에 용이

```python
plt.boxplot(y)

plt.show()
```

```python
# Extra: Plot 의 title 을 "Box plot of y"

plt.boxplot(y)

###
plt.title("Box plot of y")
###

plt.show()
```

```python
# 여러 변수들을 동시에 적용 가능

plt.boxplot((x, y))
plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프 (Bar Plot)
<br/>

- <code>.bar()</code>
- 범주형 데이터의 "값" 과 그 값의 "크기" 를 직사각형으로 나타낸 그림
- 다른 그래프에 비해서 데이터의 **도수** 확인에 용이

```python
plt.bar(x, y)

plt.show()
```

```python
# Extra: xtics 를 올바르게 처리해봅시다.

plt.bar(x, y)

###
plt.xticks(np.arange(0, 20, 1))
###

plt.show()
```

### cf) Histogram  
<br/>

- <code>.hist()</code>
- 도수분포를 직사각형의 막대 형태로 나타내는 것은 동일함
- 막대 그래프와의 차이점은 **계급**이라는 개념이 들어감
- 0, 1, 2 가 아니라 0~2 까지의 **범주형** 데이터로 구성 후 그래프를 그림
- 막대들이 연속적으로 붙어서 그려짐
- 다른 그래프에 비해서 데이터의 **도수** 확인에 용이

```python
plt.hist(y, bins = np.arange(0, 20, 2))

plt.show()
```

```python
# Extra: xtics 를 올바르게 고쳐봅시다.

plt.hist(y, bins = np.arange(0, 20, 2))

###
plt.xticks(np.arange(0, 20, 2))
###

plt.show()
```

### 원형 그래프 (Pie Chart)
- <code>.pie()</code>
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 데이터의 **비율** 확인에 용이

```python
z = [100, 300, 200, 400]

plt.pie(z)

plt.show()
```

```python
# label 을 붙이는 법

z = [100, 300, 200, 400]

###
plt.pie(z, labels = ['one', 'two', 'three', 'four'])
###

plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### Matplotlib 를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널밀도그림
- 카운트그림
- 캣그림
- 스트립그림
- 히트맵

### Seaborn Import 하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)
<br/>

- <code>sns.kdeplot()</code>
- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림

```python
# in Histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins = x)
plt.show()
```

```python
# kdeplot

sns.kdeplot(y)

plt.show()
```

```python
# 그래프에 음영을 만들어 줌

sns.kdeplot(y, shade = True) # default = False

plt.show()
```

### 카운트그림 (Count Plot)
<br/>

- <code>sns.countplot()</code>
- 범주형 column 의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과

```python
vote_df = pd.DataFrame({"name": ['Andy', 'Bob', 'Cat'], "vote": [True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False, True] , height = vote_count['name'])

plt.show()
```

```python
# sns 의 countplot

sns.countplot(x = vote_df['vote'])

plt.show()
```

### 캣그림 (Cat Plot)
<br/>

- <code>sns.catplot()</code>
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- 복잡한 데이터에 적용하기 좋음

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid.head(5)
```

```python
sns.catplot(x = "WHO Region", y = "Confirmed", data = covid)

plt.show()
```

```python
# plot 크기 바꿔주기

s = sns.catplot(x = "WHO Region", y = "Confirmed", data = covid)

###
s.fig.set_size_inches(10, 6)
###

plt.show()
```

```python
# catplot 은 어떤 그래프의 형태라기 보단 여러 개의 데이터를 묶는 그래프의 방법이고, 이 그래프의 형태를 'kind' 를 통해 설정 가능

s = sns.catplot(x = "WHO Region", y = "Confirmed", data = covid, kind = 'violin') # default = 'strip'

s.fig.set_size_inches(10, 6)

plt.show()
```

### 스트립그림 (Strip Plot)
<br/>

- <code>sns.stripplot()</code>
- scatter plot 과 유사하게 데이터의 수치를 표현하는 그래프

```python
sns.stripplot(x = "WHO Region", y = "Recovered", data = covid)

plt.show()
```

### cf) swarmplot
<br/>

- <code>sns.swarmplot()</code>
- stripplot 과는 다르게 점들이 겹치는 경우, 점들이 양 옆으로 분산시켜줌

```python
sns.swarmplot(x = "WHO Region", y = "Recovered", data = covid)

plt.show()
```

### 히트맵 (Heatmap)
<br/>

- <code>sns.heatmap()</code>
- 데이터의 행렬을 색상으로 표현해주는 그래프
- 상관계수행렬에서 자주 쓰임

```python
# 히트맵 예제

covid.corr()
```

```python
# 밝은 부분은 양의 상관관계, 어두운 부분은 음의 상관관계 의 정도를 알려줌

sns.heatmap(covid.corr())

plt.show()
```
