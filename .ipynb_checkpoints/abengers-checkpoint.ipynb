```python
df.Alignment.unique()
```

```python
def draw_graph(col_name):
    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    sns.kdeplot(x=df[df.Alignment == 'good'][col_name], ax=ax)
    sns.kdeplot(x=df[df.Alignment == 'neutral'][col_name], ax=ax)
    sns.kdeplot(x=df[df.Alignment == 'bad'][col_name], ax=ax)
    sns.kdeplot(x=df[df.Alignment.isnull()][col_name], ax=ax)
    plt.title(str(col_name))
    plt.legend(['Good', 'Neutral', 'Bad', 'ETC'])
    plt.show()
```

```python
columns = df.columns
columns[2:]
```

```python
for col in columns[2:]:
    draw_graph(col)
```
