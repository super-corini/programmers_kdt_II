```python
sin_country_list = []
for i in country_list:
        for j in i:
            if j ==',':
                break
        else:
            sin_country_list.append(i)
sin_country_list
# 정확한 결과인지에 대한 확인을 어떻게해야할지 모르겠어서
# 일단 일일이 비교해보면서 확인은해봤는데
# 뭔가 좋은방법이 떠오르지않습니다
```

### what is 5-MA ?
-  주식을 매매하는데 있어 기술적 지표로 이용하는 단위
-  최근 거래일 5일 동안의 **종가**의 평균값을 선으로 연결해 놓은 것
-  10일선, 20일선, 120일선으로 나타내기도함

```python
import matplotlib.pyplot as plt

bitcoin = pd.read_csv("./Bitcoin.csv")
bitcoin.sort_values(by = "Date",inplace = True)
bitcoin = bitcoin[bitcoin["Date"] > "2016-05-27"]
bitcoin = bitcoin[bitcoin["Date"] < "2017-07"]
bitcoin
```

```python
x = bitcoin[['Date','Close']]
x.tail(377)
```

```python
close_price = bitcoin['Close']
close_price
```

```python
print(close_price.iloc[0:5].mean())
close_price.iloc[0:5]
```

```python
mafive = []
for i in range(377):
    mafive.append(close_price.iloc[i:i+5].mean())
mafive
```

```python
date_list = []
for i in range(377):
    date_list.append(bitcoin['Date'].iloc[i+4])
```

```python
date_list
```

```python
atemp=[]
for i in range(6,19):
    for j in range(1,32):
        if i > 12:
            y= '2017-'
            if i - 12 < 10:
                m = '0'+str(i-12)+'-'
            else:
                m = str(i-12)+'-'
        else:
            y ='2016-'
            if i < 10:
                m = '0'+str(i)+'-'
            else:
                m = str(i)+'-'
        if j < 10:
            d= '0'+str(j)
        else:
            d= str(j)
        a= y+m+d
        atemp.append(a)
set(atemp)-set(date_list)
                
# 이더리움 계산할려고 날짜 세보니 다르다는걸 깨닫고 
# 2017 년 이후로 매달 10~12일 데이터가 없다는걸 확인
# 빈 날짜를 전날(9일로) 통일하여 정하기로 결정
```

```python
plt.plot(date_list,mafive,"#f2a900")
plt.plot(figsize=(20,20))
plt.xlabel("Date")
plt.ylabel("Price($)")
plt.title("Bitcoin 5-MA(Moving Average)")
plt.xticks(rotation = 90)
plt.xticks(np.arange(0, 377,28.99), labels=['2016 Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', '2017 Jan', 'Feb', 'Mar', 'Apr', 'May','June','Jul'])
# 월별로 날짜수가 달라서 이렇게 하면 안되는걸 알지만 ㅠㅠ
plt.show()
```
