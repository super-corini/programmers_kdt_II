```python
trace1 = go.Box(x=d['gender'], y=d['math score'])
data = [trace1]
layout = go.Layout(title='성별에 따른 수학점수', xaxis_title='성별', yaxis_title='수학 점수')
fig = go.Figure(data=data, layout=layout)

csp.plot(fig)

fig.to_json()
# 남성이 여성보다 수학점수가 상대적으로 높다.
```

```python
trace2 = go.Box(x=d['gender'], y=d['writing score'])
data = [trace2]
layout = go.Layout(title='성별에 따른 글쓰기점수', xaxis_title='성별', yaxis_title='글쓰기 점수')
fig = go.Figure(data=data, layout=layout)

csp.plot(fig)

fig.to_json()
# 남성이 여성보다 쓰기점수가 상대적으로 높다.
```

```python
trace3 = go.Box(x=d['gender'], y=d['reading score'])
data = [trace3]
layout = go.Layout(title='성별에 따른 읽기점수', xaxis_title='성별', yaxis_title='읽기 점수')
fig = go.Figure(data=data, layout=layout)

# csp.plot(fig)

fig.to_json()


# 남성이 여성보다 읽기점수가 상대적으로 높다.
```

```python
d['mean'] = (d['math score'] + d['reading score'] + d['writing score'])/3
```

```python
trace4 = go.Box(x = d['parental level of education'], y=d['mean'])
data = [trace4]
layout =  go.Layout(title='학위에 따른 평균점수', xaxis_title='학위', yaxis_title='평균 점수')
fig = go.Figure(data=data, layout=layout)
# fig.update_xaxes(['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"])
# csp.plot(fig)

fig.to_json()

# 학위가 높을수록 점수가 높다.
```

```python
trace5 = go.Violin(x=d['lunch'], y=d['mean'])
data = [trace5]
layout = go.Layout(title='점심식사 여부에 따른 평균점수', xaxis_title='점심식사 여부', yaxis_title='평균 점수')
fig = go.Figure(data=data, layout=layout)
# csp.plot(fig)
fig.to_json()

# 점심을 먹고온 사람들의 평균점수가 높다.
```

- 1. 성별에 따른 점수

    - 평균적으로 남자보다 여자는 수학점수가 높다.

- 2. 최종학위에 따른 평균점수

    - x축 : degree(학위)
    - 학위 순서 
        - 'some high school' < 'high school' < 'some college' < "associate's degree" < "bachelor's degree" < "master's degree"

    - 최종 학위가 더 높을수록 평균점수가 점점 더 높아진다.
