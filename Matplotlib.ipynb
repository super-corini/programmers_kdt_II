## Matplotlib Keywords

### Basics
- `plt.plot()` : draws a line graph (꺽은선그래프), default = argument: y-value, index: x-value
- `plt.show()` : show the plotted graph 

### Axis
- `np.arange(start,end,interval)` : start ~ end까지 interval 간격으로 수를 만듦 (x값을 촘촘히 할때 사용됨)
- `plt.axis([x_min,x_max,y_min,y_max])` : x,y slicing
- `plt.xticks([])` : x축 눈금설정 -> 주로 안에 반복문 설정 (i for i in range(s,e,interval)), e는 포함 안되는 것에 유의
- `plt.yticks([])` : y축 눈금설정 


### Size
- `plt.figure` : empty figure with no axes (빈도면을 선언)
- `plt.figure(figsize=(h,v))` : creates a h*v(pixel) figure 


### Labels
- `plt.title("")` : title of the graph
- `plt.legend()` : 선이 의미하는 바를 명시, .plot과 같은 그래프 그리는 함수의 두번째 인자로 주어지는 "label"의 값과 동일 (must use after the graph has been plotted)
- `plt.xlabel("")` : name of x
- `plt.ylabel("")` : name of y

```python
plt.plot([1,2,3,4,5])  # 실제 plotting을 하는 함수 -> y = [1,2,3,4,5] (인자값), x = [0,1,2,3,4] (인덱스값)
plt.show()  #plt를 확인하는 명령 

# y = x+1 함수가 결과적으로 완성됨
```

```python
plt.figure(figsize=(6,6))             # the size of the figure changes
plt.plot([1,2,3,4,5],[2,4,6,8,10])    # y = 2x
plt.show()
```

#### Creating a Quadratic Function 

```python
# y = x^2

# 점찍기 방식 이용
x = np.array([1,2,3,4,5])     # 정의역
y = np.array([1,4,9,16,25])     # f(x)

plt.plot(x,y)
plt.show()

# 문제점 ==> 뚝뚝 끊긴 선. 
# 해결책 ==> 부드러운 곡선을 만들기 위해 x값을 더 잘게 쪼개기
```

```python
# np.arange(start,end,interval) : interval 소숫점도 가능, end는 포함안됨

x = np.arange(-10,10,0.01)
plt.plot(x, x**2)      # ** : 제곱을 의미
plt.show()
```

```python
## x와 y label 만들기
x = np.arange(-10,10,0.01)

plt.xlabel("x value")
plt.ylabel("y value")

plt.plot(x, x**2)      
plt.show()
```

```python
## x,y 범위 자르기
x = np.arange(-10,10,0.01)

plt.axis([-5,5,0,25])
plt.xlabel("x value")
plt.ylabel("y value")

plt.plot(x, x**2)      
plt.show()
```

```python
## x,y 눈금 설정

x = np.arange(-10,10,0.01)

plt.axis([-5,5,0,25])
plt.xlabel("x value")
plt.ylabel("y value")

plt.xticks([i for i in range(-5,6,1)]) # x축 눈금 설정 => -5,-4,-3,-2 ... (주의: 실제범위는 4까지임, 2번째인자)
plt.yticks([i for i in range(0,26,3)]) # y축 눈금 설정 => 0,3,6,9 ...

plt.plot(x, x**2)      
plt.show()
```

```python
## title 설정, 범례설정
x = np.arange(-10,10,0.01)

plt.axis([-5,5,0,25])
plt.xlabel("x value")
plt.ylabel("y value")
plt.xticks([i for i in range(-5,6,1)]) 
plt.yticks([i for i in range(0,26,3)]) 

plt.title("y = x^2 graph")

plt.plot(x, x**2, label="trend")  
plt.legend()

plt.show()
```

## Matplotlib Graphs

### Plot : line graph
- `plt.plot(x,y)` : draws a line graph, 꺽은선그래프
- 쓰임 )) 시계열에 주로 사용됨, x축이 시간인 경우 : 시간에 따른 값의 변화를 쉽게 확인할 수 있음

### Scatter Plot 
- `plt.scatter(x,y)` : 산점도
- 쓰임 )) x와 y의 상관관계를 파악할 때 주로 사용됨 : 서로 다른 data가 연관성이 있는지 없는지 여부를 파악

### Box Plot
- `plt.boxplot((x,y))` : shows Q1,Q2(=median),Q3,min,max of a single data (주의: 인자는 하나만 전달가능, 두개일 경우 tuple 형태사용)
- 쓰임 )) data값의 분포도, 범위를 파악하는데 주로 사용됨. 

### Bar Plot
- `plt.barplot()` : 막대그래프, bar graph
- 쓰임 )) 

### Histogram
- `plt.hist(y, bins=np.arange(start,end,interval)` : 히스토그램, histogram
- 쓰임 )) x축이 하나의 특정한값이 아니라 범주(range)임, 계급이라고도 불림.

### Pie Chart
- `plt.pie(data)` : 원형그래프, 각 값의 label이 있는 것이 추천됨
- 쓰임 )) 주로 전체에 대한 부분의 비율을 한눈에 파악하기에 용이

```python
## plot
x = np.arange(20) #0~19
y = np.random.randint(0,20,20) # creates a random number btw 0~20, 20 times

plt.axis([0,20,0,20])
plt.yticks([0,5,10,15,20])
plt.xticks([i for i in range(21)])

plt.plot(x,y)
plt.show()
```

```python
## scatter plot
plt.scatter(x,y)
plt.show()
```

```python
## box
plt.boxplot((x, y))
plt.title("Box plot of x & y")
plt.show()
```

```python
## bar graph
plt.bar(x,y)
plt.xticks(np.arange(0,20,1))
plt.show()
```

```python
## histogram
plt.hist(y, bins=np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

```python
## pie chart
plt.pie([100,300,200,400], labels=["one","two","three","four"]) #pie의 경우 약간필수
plt.show()
```
