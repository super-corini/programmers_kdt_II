#### Pandas 활용 4. 행을 기준으로 데이터 접근하기

```python
# 예시 데이터 - 도서관 정보

books_dict = {"Available" :[True, True, False], "Location":[102, 215, 323], "Genre":["Programming", "Physics", "Math"]}

books_df = pd.DataFrame(books_dict, index=['버그란 무엇인가','두근두근 물리학','미분해줘 홈즈'])

books_df
```

#### 인덱스를 이용해 가져오기 : `.loc[row,col]`

```python
books_df.loc["버그란 무엇인가"]
```

```python
# "미분해줘 홈즈 책이 대출 가능한가?"

books_df.loc["미분해줘 홈즈", "Available"]
```

숫자 인덱스를 이용해 가져오기 : `.iloc[rowidx, colidx]`

```python
# 인덱스 0행의 인덱스 1열 가지고 오기

books_df.iloc[0,1]
```

```python
# 인덱스 1행의 인덱스 0~1열 가지고 오기

books_df.iloc[1, 0:2]
```

#### Pandas 활용 5. groupby

- Split : 특정한 '기준'을 바탕으로 DataFrame을 분할
- Apply : 통계함수 - sum(), mean(), median(), ... 을 적용해 각 데이터를 압축
- Combine : Apply된 결과를 바탕으로 새로운 Series를 생성 ( group_key : applied_value )

`.groupby()`

```python
# WHO Region 별 확진자수

# 1. covid에서 확진자 수 column만 추출한다.
# 2. 이를 covid의 WHO Region을 기준으로 groupby한다.

covid_by_region = covid['Confirmed'].groupby(by=covid["WHO Region"])
```

```python
covid_by_region.sum()
```

```python
# 국가당 감염자 수 

covid_by_region.mean() # sum() / 국가 수
```

## 2. covid 데이터에서 신규 확진자가 없는 나라 중 WHO Region이 'Europe'를 모두 출력하면?

Hint : 한 줄에 동시에 두 가지 조건을 Apply하는 경우 Warning이 발생할 수 있다.

```python
# 1. 신규 확진자가 없는 조건
no_new_cases = covid[covid['New cases']==0]

# 2. WHO Region이 'Europe'인 경우
no_new_cases[no_new_cases['WHO Region']=='Europe']
```

```python
# Hint의 경우.. 두 가지 조건을 Apply하는 경우 Warning이 발생하는 것을 확인할 수 있다.

covid[covid['New cases']==0][covid['WHO Region']=='Europe']
```
