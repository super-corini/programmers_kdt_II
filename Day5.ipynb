```python
import numpy as np
import pandas as pd

stat=pd.read_csv('./charcters_stats.csv')
stat
```

1. 캐릭터는 저마다 지능, 힘, ... 등 다양한 수치를 지니고 있다. 이러한 수치의 합이 가장 큰 캐릭터는 누구인가? 이를 보이기 위한 과정을 보여라.

```python
#total이 가장 큰 캐릭터 찾기
total_max=stat['Total'].idxmax(axis=0,skipna=True)
stat.loc[total_max]
```

2. 좋은 캐릭터와 나쁜 캐릭터의 능력치들의 수치 분포를 알고 싶다. 이를 표현하기 위한 적절한 그래프를 선택해서 이를 위한 전처리를 진행하고, 시각화하여라.

```python
#Alignment=good/bad를 나눠서 능력치 보기
import matplotlib.pyplot as plt
import seaborn as sns
#1.Intelligence
fig,ax=plt.subplots(1,1,figsize=(10,5))
sns.kdeplot(x=stat[stat.Alignment=='good']['Intelligence'],ax=ax)
sns.kdeplot(x=stat[stat.Alignment=='bad']['Intelligence'],ax=ax)

plt.legend(['good','bad'])
plt.show()
```
