### 가족이 있으면 살 확률이 높을까?
- 가족이 0일때, 죽은 사람의 수가 산 사람의 수보다 2배이상 차이가 남.
- 가족이 1~3명 까지 일 때, 산 사람의 수가 더 높았음.
- 가족이 4명 이상부터 죽은 수가 더 많다

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(titanic_df['SibSp']+titanic_df['Parch'],hue='Survived',data=titanic_df)
plt.subplot(1,2,2)
sns.barplot(x=titanic_df['SibSp']+titanic_df['Parch'],y='Survived',data=titanic_df)
plt.show()
```

### 그럼 가족이 많으면 Pclass는 어떨까? 

- class1과 class2에선 가족 수가 증가할수록 점점 줄어들음.
- class3도 가족 수가 증가할수록 줄어들지만, class1과 2와는 다르게 4명 이상에도 꾸준히 존재함.

### 또한 가족수와 Pclass를 기준으로 생존률은?
- class1에선 가족의 수가 증가할수록 생존률이 높았으나, 5명 이상일땐 생존률이 절반으로 떨어짐
- class2에선 가족의 수가 증가할수록 생존률이 높아짐.
- class3도 가족의 수가 증가할 수록 생존률이 높았으나, class1과 2에 비해 생존률이 낮았고, 4명 이상의 가족들의 생존률은 0에 가까움.

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(titanic_df['SibSp']+titanic_df['Parch'],hue='Pclass',data=titanic_df)
plt.subplot(1,2,2)
sns.barplot(x=titanic_df['SibSp']+titanic_df['Parch'],y='Survived', hue='Pclass',data=titanic_df)
plt.show()
```

### 1. 혈중 평균 포도당 수준이 높으면 심장병이 있을까?

- 인과 x, 상관 o
- 대체적으로 평균 포도당 수치가 약 125이상부터 심장병 질환과 상관있어 보인다.

```python
fig, ax =plt.subplots(1,1,figsize=(10,5))
sns.kdeplot(data=heart_disease_df[heart_disease_df.heart_disease==1]['avg_glucose_level'],ax=ax)
sns.kdeplot(data=heart_disease_df[heart_disease_df.heart_disease==0]['avg_glucose_level'],ax=ax)
plt.legend(["disease","not disease"])
plt.show()
```

### 2. 거주 유형과 심장병과는 상관이 있을까?

- 도심이나 농촌이나 심장병에 걸린 사람수는 같음

```python
heart_disease_df.groupby(['heart_disease','Residence_type']).count()

heart_disease_df.groupby(['heart_disease','Residence_type'])['heart_disease'].count()
```

```python
sns.catplot(x='Residence_type', col='heart_disease', kind='count',data=heart_disease_df)
plt.show()
```

### 3. 심장병은 성별과 나이에 따라 다를까?

- 그래프로 봤을 때, 심장병은 40대 이후로 급격하게 증가하는 양상을 보임
- 남,여 모두 40~80세까지 심장병 걸릴 확률이 높고 그 이후 급격히 하락(80세 이후의 표본이 적을 것이라 판단)
- 약 45~75세까지 여성보다 남성이 심장병 걸린 비율이 높으며 75세부터는 여성의 심장병 비율이 남성보다 높음

```python
fig, ax =plt.subplots(1,1,figsize=(10,5))

heart_disease_1=heart_disease_df[heart_disease_df.heart_disease==1]
heart_disease_1_M=heart_disease_1[heart_disease_1.gender=='Male']['age']
heart_disease_1_F=heart_disease_1[heart_disease_1.gender=='Female']['age']
sns.kdeplot(data=heart_disease_1_M,ax=ax)
sns.kdeplot(data=heart_disease_1_F,ax=ax)
plt.legend(["Man with heart disease","Woman with heart disease"])
plt.show()
```
