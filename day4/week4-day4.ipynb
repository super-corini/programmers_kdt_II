```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
card_df = pd.read_csv('./Credit Card Customer Data.csv')
```

```python
card_df
```

```python
## 가설: 온라인 방문횟수가 높을수록 은행 방문횟수와 전화 횟수는 감소한다

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

sns.pointplot(x=card_df['Total_visits_online'], y=card_df['Total_visits_bank'],ax=ax[0])
sns.pointplot(x=card_df['Total_visits_online'], y=card_df['Total_calls_made'],ax=ax[1])

plt.show()
```

### 결과

1. 온라인 방문횟수가 높을수록 은행 방문횟수가 적어지는 가설은 맞으나
2. 온라인 방문횟수가 적은 초기에는 오히려 전화 문의 횟수가 높아짐: 온라인 사용법 미숙으로 인한 문의전화 등으로 추정됨
