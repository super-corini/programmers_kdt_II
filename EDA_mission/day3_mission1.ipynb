```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
titanic_df = pd.read_csv('./train.csv')
titanic_df
```

```python
## 상관계수
titanic_df.corr()
```

```python
# 형제가 많을수록, 부모가 모두 타이타닉 호에 탑승했을 것.

titanic_df[['Parch', 'SibSp']].groupby(['SibSp']).mean()
```

```python
sns.catplot(x='SibSp', y='Parch', kind='point', data=titanic_df)
plt.show()
```
