### 2. 능력 수치 분포 시각화

- rader plot이 적절하다. 이는 극좌표로 플로팅 해야한다.

```python
df = df.sort_values('Total', ascending=False)
```

```python
labels = df.columns[2:-1].tolist()
labels.append(labels[0])
angles = np.linspace(0, np.pi * 2, len(labels))

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(1,1,1, polar=True)
ax.set_xticks(angles)
ax.set_xticklabels(labels)
ax.tick_params(axis='x', pad=16)
ax.set_ylim(0,100)
for i in range(0,12,4):
    ax.plot(angles, df.iloc[i][labels], label=df.iloc[i]['Name'])
    ax.fill(angles, df.iloc[i][labels], alpha=0.15)
ax.legend(bbox_to_anchor=(0.1,0.9))
plt.show()
```

```python
def rader_plot(char_idx):
    labels = df.columns[2:-1].tolist()
    labels.append(labels[0])
    angles = np.linspace(0, np.ip * 2, len(labels))

    fig = plt.figure(figsize=(10, 10))
    ax = plt.subplot(1,1,1, polar=True)
    ax.set_xticks(angles)
    ax.set_xticklabels(labels)
    ax.tick_params(axis='x', pad=16)
    ax.set_ylim(0,100)
    ax.plot(angles, df.loc[char_idx][labels], label=df.loc[char_idx]['Name'])
    ax.fill(angles, df.loc[char_idx][labels], alpha=0.15)
    ax.legend(bbox_to_anchor=(0.1,0.9))

    return fig,ax

```
